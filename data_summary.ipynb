```python
consum = pd.read_csv('data/25.세종시_가구_월평균_소비지출액.csv')
```

```python
consum
```

```python
def consumType(column_name):
    df = consum[consum['중분류']==column_name]
    df = df.drop(columns=['중분류']).rename(columns={'소분류':column_name})
    df = df.set_index(column_name)
    df = df.rename(mapper=lambda x: x.replace('만원 미만', ''),  axis='columns')
    return df



consum_age = consumType('연령별')
```

```python
consum_area = consumType('읍·면·동별')
consum_area
```

```python
consum_area.transpose().plot()
```

```python
consum_sex = consumType('성별')
consum_sex
```

```python
consum_sex.transpose().plot()
```
