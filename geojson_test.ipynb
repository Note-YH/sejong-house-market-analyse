```python
import pandas as pd
import geopandas as gpd

```

```python
geo_age = gpd.read_file('data/19.세종시_연령별_거주인구정보_격자.geojson')
```

```python
df_cnty = pd.read_csv('data/preprocess/행정동통계(2020).csv')
geo_path = 'data/32.세종시_행정경계(읍면동).geojson'
```

```python
df_cnty = df_cnty.rename(columns={'Unnamed: 0':'행정동'})
df_cnty.head(3)
```

```python
import folium
center = [36.4792025,127.288442 ]
m = folium.Map(location=center, zoom_start=10)

folium.Choropleth(
    geo_data=geo_path,
    data = df_cnty,
    columns=('행정동', '총인구수'),
    key_on = 'properties.ADM_DR_NM',
    fill_color='BuPu',
    legend='행정동 인구수'

).add_to(m)
m
```
